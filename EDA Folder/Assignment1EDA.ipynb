```python
#What is the mean number of shares after removing the outliers?


# Step 1: Calculate Q1, Q3, and IQR
Q1 = df[" shares"].quantile(0.25)  # 25th percentile
Q3 = df[" shares"].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile Range

# Step 2: Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Filter out outliers
df_filtered = df[(df[" shares"] >= lower_bound) & (df[" shares"] <= upper_bound)]
df_filtered.head()
# Step 4: Calculate mean after removing outliers
mean_filtered = df_filtered[" shares"].mean()

print(f"Mean number of shares after removing outliers: {mean_filtered}")



```

```python
#What is the approximate standard deviation of the shares after removing the outliers?

df_filtered[" shares"].std()
```

```python
gr=pd.read_csv("grades.csv")
gr.head()



gr['Is_Zip'] = gr['submission'].str.endswith('.zip')


gr['Date'] = gr['submit_time'].str.split('-').str[0]

gr['submit_time'] = pd.to_datetime(gr['submit_time'], format='%m/%d/%y-%H:%M:%S')

# Step 2: Extract the hour
gr['Hour'] = gr['submit_time'].dt.hour

df_hour = gr.groupby('Hour', as_index=False).size()

df_hour


import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df_hour, bins=24, edgecolor='black', alpha=0.7)
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Submissions')
plt.title('Distribution of Submissions by Hour')
plt.xticks(range(24))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



















```
