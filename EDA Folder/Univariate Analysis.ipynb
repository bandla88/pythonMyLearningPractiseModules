<b>Univariate analysis is the simplest form of data analysis that involves examining one variable at a time. It helps summarize and describe the characteristics of that variable using statistical methods, visualizations, and summary metrics.

Types of Univariate Analysis
Descriptive Statistics – Measures used to summarize the variable:

Measures of Central Tendency: Mean, Median, Mode
Measures of Dispersion: Range, Variance, Standard Deviation, Interquartile Range (IQR)
Skewness & Kurtosis: Shape of the data distribution
Visualization Techniques – Graphical methods to understand distribution:

Histogram: Shows frequency distribution
Boxplot: Displays median, quartiles, and outliers
Density Plot: Smooth estimate of the distribution
Bar Chart: Used for categorical variables
Probability Distributions – Identifying the nature of data:

Normal distribution, skewed distribution, uniform distribution, etc.
Example
Suppose we analyze a dataset containing the ages of employees in a company. A univariate analysis would involve:

Finding the mean age
Checking the distribution of ages using a histogram
Identifying outliers using a boxplot</b>

````python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data (ages of employees)
np.random.seed(42)
ages = np.random.normal(loc=35, scale=10, size=200)  # Mean=35, Std=10, 200 samples
ages = np.clip(ages, 18, 60)  # Clipping ages between 18 and 60

# Convert to DataFrame
df = pd.DataFrame({'Age': ages})

# Descriptive Statistics
print("Descriptive Statistics:")
print(df['Age'].describe())

# Plot Histogram
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=15, kde=True, color='blue')
plt.title('Age Distribution - Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Age'], color='orange')
plt.title('Age Distribution - Boxplot')
plt.xlabel('Age')
plt.show()
````

**Univariate Analysis: A Complete Guide with Examples**

### **What is Univariate Analysis?**
Univariate analysis is the simplest form of data analysis, involving only one variable. The primary goal is to describe, summarize, and visualize the data to understand its distribution, central tendency, and dispersion.

### **Types of Univariate Analysis**
1. **Descriptive Statistics** - Measures that summarize data.
2. **Visualization Techniques** - Graphical methods to represent data distribution.
3. **Categorical Analysis** - Frequency and proportion analysis of categorical variables.

---

### **1. Descriptive Statistics**
These measures help summarize data:

- **Mean**: The average value.
- **Median**: The middle value.
- **Mode**: The most frequent value.
- **Variance**: Measures spread of data.
- **Standard Deviation**: Square root of variance.
- **Range**: Difference between maximum and minimum.

#### **Example:**
```python
import numpy as np
import pandas as pd

# Generate synthetic data
np.random.seed(42)
ages = np.random.normal(loc=35, scale=10, size=200)
ages = np.clip(ages, 18, 60)

# Convert to DataFrame
df = pd.DataFrame({'Age': ages})

# Descriptive statistics
print(df['Age'].describe())
```

---

### **2. Visualization Techniques**
Graphical representations help understand data distribution.

#### **Histogram** (Shows frequency distribution)
```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Age'], bins=15, kde=True, color='blue')
plt.title('Age Distribution - Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### **Boxplot** (Shows outliers, median, and quartiles)
```python
sns.boxplot(x=df['Age'], color='orange')
plt.title('Age Distribution - Boxplot')
plt.xlabel('Age')
plt.show()
```

---

### **3. Categorical Analysis using `qcut`**
`qcut` is used to categorize numerical data into equal-sized quantiles.

#### **Example:**
```python
# Categorizing ages into quartiles
df['Age_Group'] = pd.qcut(df['Age'], q=4, labels=['Young', 'Mid-Young', 'Mid-Old', 'Old'])
print(df['Age_Group'].value_counts())
```

---

### **4. Pair Plot for Bivariate Insights**
Although univariate analysis focuses on a single variable, sometimes a second variable (e.g., salary) helps gain additional insights.

#### **Example:**
```python
# Generating Salary data
salaries = np.random.normal(loc=50000, scale=10000, size=200)
salaries = np.clip(salaries, 30000, 80000)
df['Salary'] = salaries

sns.pairplot(df, hue='Age_Group', palette='husl')
plt.show()
```

---

### **5. Heatmap for Correlation Analysis**
A heatmap is useful to check correlations between numerical variables.

#### **Example:**
```python
sns.heatmap(df[['Age', 'Salary']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

---

### **Summary:**
- **Univariate Analysis** focuses on one variable at a time.
- **Descriptive statistics** help summarize data.
- **Visualization** (histogram, boxplot) helps understand distribution.
- **`qcut`** is used to categorize continuous variables.
- **Pair plots** and **heatmaps** provide additional insights.

Understanding univariate analysis is the first step in exploratory data analysis (EDA) and helps prepare for more advanced techniques like bivariate and multivariate analysis!
